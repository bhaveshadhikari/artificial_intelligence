# Implement simple linear regression machine learning model.
Linear regression is the simple machine learning technique that uses statistical modeling to find realtionship between variables.

A simple linear regression model fits a straight line $ y=mx+b $  that best describes the effect of independant variable '$x$' on dependent variable '$y$'. 

import libraries: numpy for array manipulation and matplotlib for dataset visualization.

```python
import numpy as np
import matplotlib.pyplot as plt
```

initialize training data

$X$ represents successive years of an execution

$Y$ represents units sold (in thousands)

```python
X = np.array([1,2,3,4,5])  
Y = np.array([4.70,5.20,5.10,4.60,5.64])    
```

intialize parameters

```python
m = 0
b = 0
n = len(X)
```

### Gradient Descent:
to minimize the cost function by iteratively adjusting parameters in the direction of the negative gradient, aiming to find the optimal set of parameters.
##### Squared-Error $ (E) = \dfrac{1}{n} \sum \limits _{i=0} ^{n} (Y_{i} - \hat{Y}_{i})^2 $

miminize $E$ with respect to $m$ and $b$

$\dfrac{\delta{E}} {\delta{m}} = \dfrac{-2}{n} \sum \limits_  {i=0} ^{n} X_{i} (Y_{i} - \hat{Y}_{i}) $

$\dfrac{\delta{E}} {\delta{b}} = \dfrac{-2}{n} \sum \limits_  {i=0} ^{n} (Y_{i} - \hat{Y}_{i}) $

where $\hat{Y} = mX_{i} + b $

improve parameters in each iterations by:

$m = m-\alpha\dfrac{\delta{E}}{\delta{m}}$

$b = b-\alpha\dfrac{\delta{E}}{\delta{b}}$

where $\alpha$ is the learning rate.

```python
def gradient_descent(x: np.array, y: np.array, alpha: float,epochs: int,
                     m:float, b:float,n:int ) -> np.array:
    for i in range(epochs):
        y_pred = m * x + b
        dm = (-2 / n) * sum(x * (y - y_pred))
        db = (-2/n) * sum(y-y_pred)
        m = m - alpha*dm
        b = b - alpha*db
        print(f"m: {m}, b: {b}, epoch: {i}")
    return m,b
    
```

```python
m,b = gradient_descent(X,Y,0.01,100,m,b,n)
```

##### plot the relationship 'successive years' vs 'units sold'

```python
plt.scatter(X, Y, color="red", marker="o", s=40)
y_pred = m * X + b
plt.plot(X, y_pred, color="blue")
plt.xlabel('successive year')
plt.ylabel('units sold(in thousands)')
plt.show()
```

Now, make a prediction for sixth year sales.

```python
x = 6
y = m * x + b
print(f"when x = {x}, predicted y = {y}")
```
