## K-Nearest Neighbors (KNN) Algorithm
The k-nearest neighbors algorithm, also known as KNN or k-NN, is a non-parametric, supervised learning classifier, which uses proximity to make classifications or predictions about the grouping of an individual data point. While it can be used for either regression or classification problems, it is typically used as a classification algorithm, working off the assumption that similar points can be found near one another.


importing the necessary libraries and loading a dataset.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

using the Iris dataset for this example. It is a popular dataset for machine learning, consisting of 150 samples of iris flowers, each with four features (sepal length, sepal width, petal length, and petal width) and a target indicating the species of iris. There are 3 species of iris ( iris setosca, iris virgincia, and iris versicolor). The task is to classify what type of iris flower the given input sample is, based on the combination of above four features.

```python
iris = load_iris()
X = iris.data
y = iris.target
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

Now, train the model with varying the `n_neighbors` and make y prediction for each `X_test`.
store the accuracy results in the list `accList`.

```python

accList = []
for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    accList.append(accuracy_score(y_test,y_pred))

plt.plot(range(1,40),accList)
    
    
```

Now lets see what was the accuraccy of prediction when n_neighbors is choosen as 35.

```python
knn = KNeighborsClassifier(n_neighbors=28)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
